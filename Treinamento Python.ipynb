# 6. Regressão Linear

Podemos fazer uma regressão linear no python

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
x = df.drop("PRICE", axis = 1)
y = df["PRICE"]
```

```python
x
```

```python
# Aqui, a gente divide os dados em duas amostras aleatórias, sendo que test tem 33% e train 1-test_size
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 100)
```

```python
x_train
```

```python
lm = LogisticRegression().fit(x_train, y_train) # Regressão Logística
```

```python
lm = LinearRegression().fit(x_train, y_train) # Regressão Linear
```

```python
lm.coef_
```

```python
coefs = pd.DataFrame(lm.coef_, x.columns, columns = ["Coeficientes"])
coefs # Mostrar os coeficientes de um jeito mais bonito
```

```python
pred = lm.predict(x_test) # Calcular os dados preditos pelo modelo
pred
```

```python
sns.scatterplot(x = pred, y = y_test, color = "black") # Dados preditos contra os reais
```

```python
np.corrcoef(pred, y_test) # Correlação entre valor predito com o valor original
```

```python
sns.distplot(y_test - pred, bins = 165) # Distribuição de Resíduos
```

```python
import statsmodels.api as sm
```

```python
x_train_lm = sm.add_constant(x_train)
lm = sm.OLS(y_train, x_train_lm).fit()
```
